# Seaborn
* Used for advanced graphs and visualization
* A complement to matplotlib
*import seaborn as sns

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# set inline 
%matplotlib inline
sns.set()
```

```python
# line plotting
# using loaddataset

new_iris = sns.load_dataset('iris') # iris dataset

# display first five rows using using the head()
new_iris.headad()
```

```python
new_iris.head(3)
```

```python
# visualize data using line plot for our loaded data set
new_iris.petal_length.plot()
```

```python
new_iris.plot()
```

# Bar Plots

```python
new_data = pd.Series(np.random.rand(7), index=list("ABCDEFG"))
```

```python
new_data
```

```python
new_data.plot.bar(color="r", alpha=1)
```

```python
new_data.plot.barh(color="g", alpha=1)
```
