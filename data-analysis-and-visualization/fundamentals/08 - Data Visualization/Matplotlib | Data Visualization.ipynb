# Matplotlib
* Using matplotlib library to visualize data with plots and graphs
* It's an open source library

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creata array of numbers
var_nums = np.arange(9)
```

```python
%matplotlib inline
```

```python
# plot a linear graph
plt.plot(var_nums)
plt.title("Sample Graph")
plt.xlabel("x label name")
plt.ylabel("y label name")
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [1, 4, 9, 16, 25, 30]
```

```python
plt.plot(x,y)
plt.savefig("sample.png")
```

```python
# bar plots
var_file = ["file1", "file2", "file3", "file4"]
awards = [3, 9, 1, 5]

plt.bar(range(len(var_file)), awards)
plt.title("The top files")
```

# Pie Charts 
* using matplotlib.pyplot to craete Pie Charts
* plt.pie()

```python
x = np.array([37, 27, 27, 17])
ch_labels = ["firts", "second", "third", "fourth"]

plt.pie(x, labels=ch_labels)
plt.show()
```
