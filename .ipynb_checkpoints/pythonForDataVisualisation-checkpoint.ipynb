# Data Visualisation 
Data visualization involves exploring data through visual representations. It’s closely associated with data mining, which uses code to explore the patterns and connections in a data set.

## Example 1.0 | Let's plot a line using matplotlib, and then customize it to create a more informative visualisation of our data

```python
import matplotlib.pyplot as plt

input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]

# Set the chat title and label axes
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels
plt.tick_params(axis="both", labelsize=14)

plt.plot(input_values, squares, linewidth=5) # Plot the values
plt.show() # Open the matplotlib viewer and display the plot
```

## Example 1.1 | Plotting and Styling Individual Points with scatter()
To plot a single point, use the ***scatter()*** function. Pass the single (x, y)
values of the point of interest to scatter(), and it should plot those values:

```python
## Using the scatter() function to plot a point
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

plt.scatter(x_values, y_values, s=100)

# Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# Set size of tick labels.
plt.tick_params(axis="both", which="major", labelsize=14)

plt.show()
```

# Example 1.2 | Calculating Data Automatically
Writing out lists by hand can be inefficient, especially when we have many points. Rather than passing our points in a list, let’s use a loop in Python to do the calculations for us. Here’s how this would look with 1000 points:

```python
x_values = list(range(1, 10001)) ##  Generate x values from 1 to 1000 and place them in x_values as a list
y_values = [x**2 for x in x_values] ## square corresponding x_values and place the result in y_values

plt.scatter(x_values, y_values, s=2) ## plot the co-ordinates to the user

## Customize our graph
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)
plt.tick_params(axis="both", which="major", labelsize=14)

## Set the range() for each axis
plt.axis([0, 1100, 0, 1100000])

plt.show() ## open the matplotlib viewer and display the plots
```
